```python
def compute_model_fun(x,w,b):
    l=x.shape[0]
    fwb=np.zeros(l)
    for i in range(l):
        fwb[i]=w*x[i]+b

    return fwb
```

Model representation and cost function

```python
def compute_model_and_cost(x,y,w,b):
    l=x.shape[0]
    fwb=np.zeros(l)
    jwb=0
    for i in range(l):
        fwb[i]=w*x[i]+b
        jwb=jwb+(fwb[i]-y[i])**2
    Jwb=(1/(2*l))*jwb    
    return Jwb,fwb
```

```python

w=202
b=100
[J_wb,f_wb]=compute_model_and_cost(xt,yt,w,b)
print(f"Model prediction f_wb is : {f_wb}")
print(f"Cost function J_wb is : {J_wb}")
```
